```python

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
import re
import string

class Preprocessor:

    def __init__(self):
        # Create a variable of stop words.
        self.stopwords = set(stopwords.words('english'))
        self.stemmer = PorterStemmer()

    def preprocess(self, text):
        text = self.remove_links(text)
        text = self.remove_punctuations(text)
        normalized_text = self.normalize(text)
        tokens = self.word_tokenize(normalized_text.lower())
        return tokens

    def normalize(self, text):
        tokens = self.word_tokenize(text.lower())
        filtered_tokens = self.remove_stopwords([token for token in tokens if token not in self.stopwords])
        lemmatizer = WordNetLemmatizer()
        normalized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
#         stemmed_tokens = [self.stemmer.stem(token) for token in filtered_tokens]
#         normalized_text = ' '.join(stemmed_tokens)
        return ' '.join(normalized_tokens)

    def remove_links(self, text):
        url_pattern = re.compile(r'http[s]?://\S+|www\.\S+')
        cleaned_text = re.sub(url_pattern, '', text)
        return cleaned_text

    def remove_punctuations(self, text):
        table = str.maketrans("","")
        return text.translate(str.maketrans('','', string.punctuation))

    def word_tokenize(self, text):
        tokens = word_tokenize(text)
        return tokens

    def remove_stopwords(self, words):
        filtered_tokens = [token for token in words if token not in self.stopwords]
        return filtered_tokens
```

<h3> دستورات </h3>
<li>متد <code>sort_based_index_construction</code> از <code>DocumentBatch</code>:</li>
<p>هر سند را با استفاده از تابع‌هایی که در قسمت قبل نوشتید، عمل preprocessing را روی آن انجام دهید.</p>
<p>برای هر توکن، شناسه سند را به فهرست معکوس (inverted index) برای آن توکن اضافه کنید.</p>
<li>متد <code>add_batch</code> در <code>FastSearchEngine</code></li>
<p>فهرست معکوس برای دسته (batch) را ایجاد کنید.</p>
<p>این دسته را به فهرست‌های روزانه اضافه کنید.</p>
<li>متد <code>end_of_day_merge</code> از <code>FastSearchEngine:</code></li>
<p>استراتژی ادغام لگاریتمی را پیاده‌سازی کنید تا به صورت بهینه فهرست‌های روزانه را با فهرست اصلی ادغام کنید.</p>

```python
from collections import defaultdict, deque
import time

class DocumentBatch:
    def __init__(self, docs):
        self.documents = docs
        self.index = defaultdict(set)

    def sort_based_index_construction(self):
        # {TODO}: Build an inverted index for the batch
        preprocessor = Preprocessor()
        for index, row in self.documents.iterrows():
            doc_tokens = preprocessor.preprocess(row['abstract'])
            doc_tokens = set(doc_tokens)
            for token in doc_tokens:
                if token in self.index:
                    self.index[token].append(row['paperId'])
                else:
                    self.index[token] = [row['paperId']]
#         print(self.index)

class FastSearchEngine:
    def __init__(self):
        self.main_index = defaultdict(set)
        self.daily_indices = deque()

    def add_batch(self, batch: DocumentBatch):
        # {TODO}: Incorporate the new batch into the daily indices
        self.daily_indices.append(batch)

    def end_of_day_logarithmic_merge(self):
        # {TODO}: Implement the logarithmic merge strategy
        while self.daily_indices:
            batch = self.daily_indices.popleft()
            for key, paperIds in batch.index.items():
                if key in self.main_index:
                    # TODO, check if paperIds not repeated before 
                    self.main_index[key] = self.main_index[key] + paperIds
                else:
                    self.main_index[key] = paperIds


num_servers = 2
num_server_batches = 5
df_length = len(data_frame)
search_engine = FastSearchEngine()
df = data_frame
for i in range(num_servers):
    server_docs = df[i * (df_length // num_servers): min((i + 1) * (df_length // num_servers), df_length) ]
    server_docs_length = len(server_docs)
    for j in range(num_server_batches):
        server_batches = server_docs[j * (server_docs_length // num_server_batches): 
                                     min((j + 1) * (server_docs_length // num_server_batches), server_docs_length)]
        document_batches = DocumentBatch(server_batches)
        start_time = time.time()
        document_batches.sort_based_index_construction()
        print("sort_based_index_construction took: ", time.time() - start_time)
        search_engine.add_batch(document_batches)
    
    start_time = time.time()
    search_engine.end_of_day_logarithmic_merge()
    print("end_of_day_logarithmic_merge took:", time.time() - start_time)
# Divide the documents into groups to distribute them between servers. For example, let's consider two servers here.
# Divide the documents of each server into batches, for instance, five batches.
# Create an index for each batch and for each server, then merge them into the main index at the end of the day.
# Repeat this process until all documents are processed.

# Usage example
# server_a_docs = ["sample_1", "sample_2", "sample_3", "sample_4", "sample_5", "sample_6", "sample_7", "sample_8", "sample_9", "sample_10"]
# server_b_docs = ["sample_11", "sample_12", "sample_13", "sample_14", "sample_15", "sample_16", "sample_17", "sample_18", "sample_19", "sample_20"]

# for i in range(5):
#     server_a_batch = DocumentBatch(server_a_docs[i*2:i*2+2])
#     server_b_batch = DocumentBatch(server_b_docs[i*2:i*2+2])

#     search_engine = FastSearchEngine()
#     search_engine.add_batch(server_a_batch)
#     search_engine.add_batch(server_b_batch)

#     # At end of day
#     search_engine.end_of_day_logarithmic_merge()

# Note that the above was just an example demonstrating the general process.
# but you can take two servers and five batches for each server as a constant and implement the process for them.
print(search_engine.main_index)
```

<h2>پاسخ سوال بالا</h2>
پاسخ‌ خود را در این سلول بنویسید.
    
    
    برای این کار چندین روش مختلف وجود دارد. بر اساس آنجه در درس خواندیم، یک روش برای این کار نگاه کردن صرفا به خود کلمه است، و پیدا کردن کلمه‌ای که نزدیک‌ترین فاصله را به آن دارد. برای پیدا کردن نزدیک‌ترین فاصله هم روش‌های مختلفی می‌باشد از جمله Levenshtein distance و longest common subsequence و Hamming distance و ... که هر کدام کاربردهای خود را در جای مخصوص دارد.
    
    روش دیگر مبتنی بر کلمه‌ی تنها نیست و با استافده از جمله و کلمه‌های دور و بر، سعی در درست کردن اشکال املایی دارد. یعنی اگر یک گلمه در جایگاهش معنی بهتری بدهد، جایگزین می‌شود یا کلمه‌ی جدید. 
    
    اما موتورهای جست و جو به این صورت عمل می‌کنند که ابتدا عملیات Stemming و Lemmatization را روی کلمه اجرا می‌کنند تا کلمه ها را به ریشه ی خود نزدیک یا تبدیل کنند.
    سپس کلمه‌های هم معنی را اضافه می‌کنند. مثلا برای ماشین کلمه های هم معنی نزدیک به ماشین مانند موتور، چرخ و ... را به نتیاج جست و حو اضافه می‌کند. در مرحله‌ی بعدی به خود پرسش کلمه‌های بعدی اضافه می‌شود. مثلا اگر در پرسشی کمله ماشین وجود داشته باشد، پرسش به پرسش‌های مرتبط مثل ماشین پر قدرت، ماشین دو درب و ... اضافه می‌شود. 
    در نهایت هم نتایج امتیاز بندی میشود و بر اساس امتیاز به کاربر نمایش داده می‌شود.

<p>

```python
from typing import List, Dict
import itertools
import string


def get_all_bigrams():
    alphabets = string.ascii_lowercase + "$"
    permutations = list(itertools.product(alphabets, repeat=2))
    permutation_list = []
    for perm in permutations:
        permutation_list.append(''.join(perm))
    permutation_list.remove("$$")
    return permutation_list

def create_bigram_index(main_index):
    """
    Creates a bigram index for the spell correction

    Returns
    -------
    dict
        A dictionary of bigrams and their occurence
    """
    # TODO: Create the bigram index here
    bigram_dict: Dict[str, List[str]] = {}
    terms = list(main_index.keys()) + list(stopwords.words('english')) + list(string.punctuation)
    bigrams = get_all_bigrams()
    for bigram in bigrams:
        if bigrams[0] == "$":
            bigram_dict[bigram] = [term for term in terms if term.startswith(bigram[1])]
        elif bigram[1] == "$":
            bigram_dict[bigram] = [term for term in terms if term.endswith(bigram[0])]
        else:
            bigram_dict[bigram] = [term for term in terms if bigram in term]

    return bigram_dict

bigram_index = create_bigram_index(search_engine.main_index)
```

```python
import math
def get_word_bigrams(word):
    bigrams = []
    for i in range(len(word) - 1):
        bigrams.append(word[i : i + 2])
    bigrams.append(bigrams[-1][1] + "$")
    bigrams.insert(0, "$" + bigrams[0][0])
    return bigrams

def get_bigram_words(bigrams, bigram_index):
    total_words = set()
    for bigram in bigrams:
        words = bigram_index[bigram]
        for word in words:
            if word not in total_words:
                total_words.add(word)
    return list(total_words)

def get_edit_distance(word1, word2):
    m = len(word1)
    n = len(word2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j] + 0.9, dp[i][j - 1] + 1.1, dp[i - 1][j - 1] + 1)

    return dp[m][n]

def find_closest_term(term, bigram_index):
    bigrams = get_word_bigrams(term)
    bigram_words = get_bigram_words(bigrams, bigram_index)

    min_edit_distance = math.inf
    closest_word = None
    for word in bigram_words:
        edit_distance = get_edit_distance(word, term)
        if edit_distance < min_edit_distance:
            min_edit_distance = edit_distance
            closest_word = word
    
    return closest_word, min_edit_distance
    
def spell_correction(query, main_index, bigram_index):
    """
    Correct the give query text, if it is misspelled

    Paramters
    ---------
    query: str
        The query text

    Returns
    ---------
    corrected_query: str
        The corrected text
    """
  # TODO: correct the query
    corrected_query = ""

    query_terms = query.lower().split()
    valid_terms = list(stopwords.words('english')) + list(main_index.keys()) + list(string.punctuation)
    for query_term in query_terms:
        if query_term in valid_terms:
            corrected_query += f" {query_term}"
        else:
            edited_word, edit_distance = find_closest_term(query_term, bigram_index)
            corrected_query += f" {edited_word}"

    return corrected_query if corrected_query[0] != " " else corrected_query[1:]

main_index = search_engine.main_index
# Example usage
user_query = "Wht is the most populr progarmming lanuage ?"
# print(list(main_index.keys()))
# find_closest_term("wht", bigram_index)
spell_correction(user_query, main_index, bigram_index)
```

<h1> Boolean Retrieval </h1>
<p>
 در این قسمت هدف طراحی یک سامانەی بازیابی اطلاعات boolean می‌باشد. 

برای این کار ابتدا پیش پردازش‌های مورد نیاز را مانند بخش قبل بر روی متون انجام دهید و در مرحله بعد ماتریس doc−term را ایجاد کنید. در نهایت کلاس BooleanRetrievalModel را تکمیل کنید که شامل توابع preprocess_query و find_siⅿiⅼar_docs است که توضیحات هرکدام در قسمت کد موجود است. هدف نهایی این است که هرگاه کوئری به تابع find_siⅿiⅼar_docs از کلاس BooleanRetrievalModel داده شود، شناسه k تا از داک‌هایی که شامل کوئری داده شده هستند برگردانده شوند.
</p>
</div>

```python
# preprocess and tokenize all documents
def preprocess_term_matrix(documents):
    paper_tokens = {}
    preprocessor = Preprocessor()
    terms = set()
    max_paper_id = -1
    for index, doc in documents.iterrows():
        paper_tokens[doc['paperId']] = preprocessor.preprocess(doc['abstract'])
        if doc['paperId'] > max_paper_id:
            max_paper_id = doc['paperId']
        for token in paper_tokens[doc['paperId']]:
            if token not in terms:
                terms.add(token)
    terms = list(terms)
    return terms, paper_tokens, max_paper_id
```

```python
def make_doc_term_matrix(documents):
    """
    Create doc_term_matrix by using documents and unique words.
    """
    terms, paper_tokens, max_paper_id = preprocess_term_matrix(documents)
    term_document_matrix = {}
    paparId_values = {}

    for paperId, tokens in paper_tokens.items():
        paparId_values[paperId] = False

    for term in terms:
        term_document_matrix[term] = dict(paparId_values)

    for paperId, tokens in paper_tokens.items():
        for token in tokens:
            term_document_matrix[token][paperId] = True
    return term_document_matrix
```

```python
class BooleanRetrievalModel:
    def __init__(self, doc_term_matrix):
        """    
        Set doc_term_matrix and initialize the model.
        """
        self.preprocessor = Preprocessor()
        self.term_document_matrix = doc_term_matrix

    def preprocess_query(self, query):
        """
        Do necessary preprocessing here before using the query to find k similar docs.
        Use methods from Preprocess section.
        """
        processed_query = self.preprocessor.preprocess(query)
        print(processed_query)
        return processed_query
    
    def find_siⅿiⅼar_docs(self, query, k=20):
        processed_query = self.preprocess_query(query)
        """
        Find k similiar documents.
        """
        similiar_docs = None
        for term in processed_query:
            if term not in self.term_document_matrix:
                return []
            
            doc_vector = self.term_document_matrix[term]
            
            if similiar_docs is None:
                similiar_docs = doc_vector
            else:
                for key, value in doc_vector.items():
                    similiar_docs[key] = similiar_docs[key] & value
        
        return [k for k, v in similiar_docs.items() if v][:k]
```

```python
def convert_number_to_gamma_code(number):
    ofsset = bin(number)[3:]
    length = "1" * len(ofsset) + "0"
    return length + ofsset
    
def convert_inverted_index_to_gamma_code(posting_list):
    byte_array = ""
    for i in range(len(posting_list)):
        if i == 0:
            byte_array = convert_number_to_gamma_code(posting_list[i])
        else:
            byte_array = byte_array + convert_number_to_gamma_code(posting_list[i] - posting_list[i - 1])
    return byte_array
            
def gamma_encode():
    gamma_index = defaultdict(set)
    for key, value in main_index.items():
        gamma_index[key] = convert_inverted_index_to_gamma_code(value)
    
    return gamma_index

def convert_number_to_variable_byte(num):
    encoded = ""
    first_bit = "1"
    while num > 0:
        remainder = num % (2 << 7)
        num = num >> 7
        byte = format((remainder & 127), "08b")
        byte = first_bit + byte[1:]
        first_bit = "0"
        encoded = byte + encoded
        
    return encoded

def convert_inverted_index_to_variable_byte(posting_list):
    byte_array = ""
    for i in range(len(posting_list)):
        if i == 0:
            byte_array = convert_number_to_variable_byte(posting_list[i])
        else:
            byte_array = byte_array + convert_number_to_variable_byte(posting_list[i] - posting_list[i - 1])
    return byte_array

def variable_byte_encode():
    variable_index = defaultdict(set)
    for key, value in main_index.items():
        variable_index[key] = convert_inverted_index_to_variable_byte(value)

    return variable_index
```

```python
import math

def convert_gamma_bits_array_to_posting_list(bit_array):
    posting_list = []
    
    while len(bit_array) > 0:
        length = bit_array.index("0")
        offset = "1" + bit_array[length + 1: 2 * length + 1]
        if not posting_list:
            posting_list.append(int(offset, 2))
        else:
            posting_list.append(posting_list[-1] + int(offset, 2))
        bit_array = bit_array[2 * length + 1:]
    
    return posting_list

def gamma_decode(gamma_index):
    main_index = {}
    
    for key, gamma_bits in gamma_index.items():
        posting_list = convert_gamma_bits_array_to_posting_list(gamma_bits)
        main_index[key] = posting_list
        
    return main_index

def convert_variable_byte_array_to_posting_list(variable_bits):
    posting_list = []
    posting_bytes = [variable_bits[i:i+8] for i in range(0, len(variable_bits), 8)]
    current_number = 0
    for byte in posting_bytes:
        if byte[0] == "1":
            current_number = (current_number << 7) + int(byte[1:], 2)
            if not posting_list:
                posting_list.append(current_number)
            else:
                posting_list.append(posting_list[-1] + current_number)
            current_number = 0
        else:
            current_number = (current_number << 7) + int(byte[1:], 2)
            
    return posting_list
            

def variable_byte_decode(variable_index):
    main_index = {}

    for key, variable_bits in variable_index.items():
        posting_list = convert_variable_byte_array_to_posting_list(variable_bits)
        main_index[key] = posting_list

    return main_index
```
